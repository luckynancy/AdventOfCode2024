```python
def part1(i, j, row, col, input):
    sum = 0

    def is_xmas(dx, dy):
        chars = []
        for k in range(4):
            try:
                chars.append(input[i + k * dx][j + k * dy])
            except :
                pass
        return 1 if ''.join(chars)== 'XMAS' else 0
        
    if j + 3 < col:
        sum += is_xmas( 0, 1) 
        if i - 3 >= 0:
            sum += is_xmas(-1, 1)  
        if i + 3 < row:
            sum += is_xmas(1, 1)  
    
    if j - 3 >= 0:
        sum += is_xmas(0, -1)  
        if i + 3 < row:
            sum += is_xmas(1, -1)  
        if i - 3 >= 0:
            sum += is_xmas(-1, -1)  
    
    if i + 3 < row:
        sum += is_xmas(1, 0)  
    
    if i - 3 >= 0:
        sum += is_xmas(-1, 0) 

    return sum

def part2(i, j, row, col, input):
    count = 0

    directions = [
        [(0, 0), (1, 1), (2, 2)],  
        [(2, 0), (1, 1), (0, 2)],  
        [(0, 2), (1, 1), (2, 0)],  
        [(2, 2), (1, 1), (0, 0)],  
    ]
    
    def diagonal(i, j, direction):
        chars = [input[i + direction[k][0]][j + direction[k][1]] for k in range(3)]
        return 1 if ''.join(chars)== 'MAS' else 0
    
    if i + 2 < row and j + 2 < col:
        for direction in directions:
            count += (diagonal(i, j, direction))
    
    return True if count == 2 else False
 
```

```python
input = []
with open('day04.txt', encoding='utf8') as f:
    for line in f:
        input.append(list(line.strip()))

partI, partII = 0, 0
row, col = len(input), len(input[0])

for i in range(row):
    for j in range(col):
        partII += part2(i, j, row, col, input)
        if input[i][j] == 'X':
            partI += part1(i, j, row, col, input)
                     
print(partI, partII)
```
