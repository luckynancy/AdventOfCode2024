```python
import re

input = {}
frequencies = set()

x = 0
txt = ''
with open('day08.txt', encoding='utf8') as f:
    for line in f:
        txt = line.strip()
        pattern = r'[a-zA-Z]|\d'
        matches = [(m.group(), m.start()) for m in re.finditer(pattern, txt)]
        for (frequency, y) in matches:
            if frequency not in frequencies:
                frequencies.add(frequency)
                input[frequency] = set()
            input[frequency].add((x,y))
        x += 1
xmax = x
ymax = len(txt)
```

```python
import itertools

antinodes = set()
def outside(pos, xmax, ymax):
    (x,y) = pos
    return True if (x < 0) or (x > xmax-1) or (y < 0) or (y > ymax-1) else False

def new_pos1(a, b, n, part1 = True):
    dx, dy = a[0] - b[0], a[1] - b[1]

    if dx == 0 :
        c0 = d0 = a[0]
    else: 
        c0 = a[0] - dx * n 
        d0 = a[0] + dx if part1 else a[0] + dx * n

    if dy == 0 :
        c1 = d1 = a[1]
    else: 
        c1 = a[1] - dy * n
        d1 = a[1] + dy if part1 else a[1] + dy * n
    
    points = set()
    if not outside((c0,c1), xmax, ymax):
        points.add((c0,c1))

    if not outside((d0,d1), xmax, ymax):
        points.add((d0,d1))
        
    return points


def new_pos2(a, b, xmax, ymax):
    new = set()
    n = 1
    any = True
    while any: 
        points = new_pos1(a, b, n, part1=False)
        if len(points) == 0 :
            any = False
        else:
            new = new.union(points)
        n += 1
    return new

partI, partII = set(), set()

for frequency in frequencies:
    pos = input[frequency]
    pairs = list(itertools.permutations(pos, 2))
    
    for (a,b) in pairs: 
        partI = partI.union(new_pos1(a,b,n=2,part1=True))
        partII = partII.union(new_pos2(a, b, xmax, ymax))

print(len(partI), len(partII))
```
